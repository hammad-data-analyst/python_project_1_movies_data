## Top 5 highest grossing movie

```python
df = df.sort_values(by=['gross'],inplace=False, ascending=False)
df
```

## Dropping any duplicates

```python
df.drop_duplicates()
# since there are no duplicates no change in data
```

## Budget high correlation
## Company high correlation

### Scatter plot with budget vs gross

```python
plt.scatter(x=df['budget'],y=df['gross'])

plt.title('Budget vs Gross Earning')

plt.xlabel('Gross Earning')

plt.ylabel('Budget for film')
plt.show()
```

```python
df.head()
```

### Plot budget vs gross using seaborn

```python
sns.regplot(x='budget',y='gross',data=df,scatter_kws={"color": "red"},line_kws={"color": "blue"})
```

### Let's start looking at corelation

```python
df.corr(method='pearson')

# here highest corelation if of budget & gross since year is not including here
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matric')

plt.xlabel('Movie Feature')

plt.ylabel('Movie Feature')

plt.show()
```

### Some Visualizations

#### Top 5 highest budget movies

```python
hbm = df.sort_values(by=['budget'],inplace=False, ascending=False).head(5)
plt.bar(hbm['name'],hbm['budget'],color='maroon',width=0.3)

plt.title('Top 5 highest budget movies')
plt.xlabel('Movie Name')
plt.ylabel('Budget of the Film')
plt.show()
```
